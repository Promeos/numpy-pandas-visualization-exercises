# Curves to graph

- y=0x+1 and label the curve "O(1)"

- y=log(x) and label the curve "O(log n)"

- y=x and label the curve "O(n)"

- y=xâˆ—log(x) and label it "O(n log n)"

- y=x2 and label it "O(n^2)"

- y=2x and label it "O(2^n)"

- y=n! and label it "O(n!)"

- y=nn and label it "O(n^n)"

```python
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [x for x in range(1, 11)]
y1 = [0*num + 1 for num in x]
y2 = [math.log(num) for num in x]
y3 = x
y4 = [num * math.log(num) for num in x]
y5 = [math.pow(num, 2) for num in x]
y6 = [math.pow(2, num) for num in x]
y7 = [math.factorial(num) for num in x]
y8 = [math.pow(num, num) for num in x]
```

```python
plt.figure(figsize=(6, 6))


plt.plot(x, y1, label="$O(1)$")
plt.plot(x, y2, label="$Olog(n)$")
plt.plot(x, y3, label="$O(n)$")
plt.plot(x, y4, label="$O(n log n)$")
plt.plot(x, y5, label="$O(n^2)$")
plt.plot(x, y6, label="$O(2^n)$")
plt.plot(x, y7, label="$O(n!)$")
plt.plot(x, y8, label="$O(n^n)$")

plt.legend(fontsize=10)
plt.title("Big O Notation", size=10)
plt.xlabel('Elements', size=10)
plt.ylabel('Operations', size=10)
plt.ylim(0, 10)
plt.tight_layout()
plt.show();
```
