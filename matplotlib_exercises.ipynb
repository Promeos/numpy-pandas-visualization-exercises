```python
import matplotlib.pyplot as plt
import math
```

### 1.
> Use matplotlib to plot the following equation:

                     y=x2âˆ’x+2

> You'll need to write the code that generates the x and y points. Add an anotation for the point 0, 0, the origin.

```python
x = [number for number in range(1,11)]
y1 = [(n**2 - n + 2) for n in x]
```

```python
plt.plot(x, y1)
plt.title('$y = x^2 - x + 2$')
plt.annotate('origin', xy=(0, 0), xytext=(0, 20), arrowprops={'facecolor': 'orange'});
```

### 2.
> Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

                y=âˆšx

                y=x3
    
                y=2x

> You can use functions from the `math` module to help implement some of the equations above.

### y=âˆšx

```python
y2 = [math.sqrt(number) for number in x]
plt.plot(x, y2)
plt.title('$\sqrt{x}$');
```

### y=x^3

```python
y3 = [math.pow(number, 3) for number in x]
plt.plot(x, y3)
plt.title('$x^3$');
```

### y=2^x

```python
y4 = [math.pow(2, number) for number in x]
plt.plot(x, y3)
plt.title('$2^x$');
```

### 3.
> Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.figure(figsize=(12, 10))

plt.subplot(221)
plt.plot(x, y1, c='firebrick')
plt.title('$y = x^2 - x + 2$', size=20)

plt.subplot(222)
plt.plot(x, y2, c='blue')
plt.title('$\sqrt{x}$', size=20)

plt.subplot(223)
plt.plot(x, y3, c='darkorange')
plt.title('$x^3$', size=20)

plt.subplot(224)
plt.plot(x, y4, c='darkolivegreen')
plt.title('$2^x$', size=20)

plt.tight_layout()
plt.show()
```

### 4.
> Comine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend.

```python
plt.figure(figsize=(12, 10))

plt.plot(x, y1, c='firebrick', label='$y = x^2 - x + 2$')
plt.plot(x, y2, c='blue', label='$\sqrt{x}$')
plt.plot(x, y3, c='darkorange', label='$x^3$')
plt.plot(x, y4, c='darkolivegreen', label='$2^x$')

plt.legend(fontsize=15)
plt.grid()
plt.show()
```
